```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
plt.figure(figsize = (6, 4))
sns.displot(df["age"])
plt.show()
```

```python
plt.figure(figsize = (5, 3))
sns.boxplot(x = "class", y = "age", palette = "pastel", data = df)
plt.show()
df.describe().transpose()
```

```python
plt.figure(figsize = (5, 3))
sns.violinplot(x = "class", y = "age", palette = "rainbow", data = df)
plt.show()
```

```python
plt.figure(figsize = (6, 4))
pq = sns.histplot(x = "fare", bins = 10,  kde = False, hue = "survived", data = df)
for i in pq.containers:
    pq.bar_label(i)
plt.show()
```

```python
plt.figure(figsize = (5, 4))
sns.scatterplot(x = "age", y = "fare", data = df)
plt.show()
```

```python
plt.figure(figsize = (5, 4))
sns.scatterplot(x = "age", y = "fare", data = df, hue = "survived")
plt.show()
```
