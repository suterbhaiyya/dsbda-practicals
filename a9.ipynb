```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
plt.figure(figsize = (5, 4))
sns.boxplot(x = "sex", y = "age", data = df, hue = "survived", palette = "viridis")
plt.show()
```

```python
plt.figure(figsize = (5, 4))
sns.violinplot(x = "sex", y = "age", data = df, hue = "survived")
plt.show()
```

```python
plt.figure(figsize = (5, 4))
sns.catplot(x = "sex", data = df, kind = "count")
```

```python
print(df.describe().transpose())
```

```python
plt.figure(figsize = (5, 4))
ad = sns.countplot(x = "sex", data = df, hue = "survived", palette = "coolwarm")
for i in ad.containers:
    ad.bar_label(i)
```
