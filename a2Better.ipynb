```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {
    "roll": range(1, 51),
    "name": ["Student" + str(i) for i in range(1, 51)],
    "marks": np.random.randint(35, 100, size = 50),
    "gender": np.random.choice(["Male", "Female"], size = 50)
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.info()
```

```python
df.to_csv("academic_performance.csv")
```

```python
df.loc[18, "marks"] = np.nan
df.loc[9, "marks"] = np.nan
df.loc[23, "marks"] = np.nan
df.loc[44, "marks"] = np.nan
df.loc[11, "roll"] = np.nan
```

```python
df.isnull().sum()
```

```python
df["marks"].fillna(df["marks"].mean(), inplace = True)
df["marks"] = df["marks"].astype(int)
df
```

```python
df.dropna(inplace = True)
df
```

```python
df.loc[10, "marks"] = 200
df.loc[34, "marks"] = -10
```

```python
sns.boxplot(data = df, x = "marks")
plt.show()
```

```python
df = df[(df["marks"] > 0) & (df["marks"] <= 100)]
df
```

```python
sns.boxplot(data = df, x = "marks")
plt.show()
```
